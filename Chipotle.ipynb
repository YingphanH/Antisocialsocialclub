```python
import pandas as pd
import plotly.express as px
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
df.head()
```

```python
df.isna().sum()
```

```python
item_orders = df.groupby('item_name')['quantity'].sum().reset_index()
top_10_items = item_orders.sort_values(by='quantity', ascending=False).head(10)
fig = px.sunburst(top_10_items, path=['item_name'], values='quantity', title='Top 10 Items by Sum of Orders')
fig.show()
```

```python
fig = px.box(df, x='item_name', y='item_price', title='Distribution of Item Prices by Item Name')
fig.update_layout(xaxis={'categoryorder':'total ascending'})
fig.show()
```

```python
choice_counts = df[df['choice_description'].notna()].groupby('item_name')['choice_description'].count().reset_index()
display(choice_counts)
```

```python
item_price_stats = df.groupby('item_name')['item_price'].agg(['max', 'min', 'mean']).reset_index()
item_price_stats['max add on value'] = item_price_stats['max'] - item_price_stats['min']
item_sales = df.groupby('item_name')['item_price'].sum().reset_index()
item_price_stats = item_price_stats.merge(item_sales, on='item_name', suffixes=('_stats', '_total_sale'))
item_price_stats = item_price_stats.rename(columns={'item_price': 'total_sales'})
item_price_stats = item_price_stats.sort_values(by='total_sales', ascending=False).reset_index(drop=True)
item_price_stats.index = item_price_stats.index + 1
display(item_price_stats)
```

```python
choice_description_counts = df['choice_description'].value_counts().reset_index()
choice_description_counts.columns = ['choice_description', 'count']
top_10_choice_descriptions = choice_description_counts.head(10)
display(top_10_choice_descriptions)
```

```python
top_10_descriptions_list = top_10_choice_descriptions['choice_description'].tolist()
items_with_top_descriptions = df[df['choice_description'].isin(top_10_descriptions_list)]
item_name_grouped_by_description = items_with_top_descriptions.groupby('item_name').size().reset_index(name='count')
item_name_grouped_by_description = item_name_grouped_by_description.sort_values(by='count', ascending=False)
display(item_name_grouped_by_description)
```
